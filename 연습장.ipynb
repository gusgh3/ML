```python
import pandas as pd
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
train = train.drop('Cabin', axis=1)
```

```python
test = test.drop('Cabin', axis=1)
```

```python
train.describe()
```

```python
test.describe()
```

```python
train.info
```

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
```

```python
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
```

```python
train.isnull().sum()
```

```python
train = train.dropna()
```

```python
test = test.dropna()
```

```python
train.info()
```

```python
target = pd.DataFrame(target)
```

```python
target = train.pop('Survived')
```

```python
cols = ['Survived', 'PassengerId']
```

```python
train.head()
```

```python
train.info()
```

```python
# o_cols = ['Sex', 'Embarked']
cols = ['Sex', 'Embarked', 'Pclass', 'Age', 'SibSp' ,'Parch', 'Fare']
```

```python
train_n = train[cols]
test_n = test[cols]
```

```python
train_n = pd.get_dummies(train_n)
test_n = pd.get_dummies(test_n)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler() 
train_scaler = ss.fit_transform(train_n)
test_scaler = ss.transform(test_n)
```

```python
train_scaler
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(train_scaler, target, test_size=0.2, random_state=1204)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, max_depth=9, random_state=1204)
rf.fit(train_input, train_target)
```

```python
A = rf.predict(test_input)
```

```python
from sklearn.metrics import accuracy_score

AA = accuracy_score(test_target, A)
print(AA)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.01, max_depth=9, random_state=1204)
gb.fit(train_input, train_target)
```

```python
B = gb.predict(test_input)
```

```python
from sklearn.metrics import accuracy_score

BB = accuracy_score(test_target, B)
print(BB)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(learning_rate=0.1, max_depth=9, random_state=1204)
hgb.fit(train_input, train_target)
```

```python
C = hgb.predict(test_input)
```

```python
from sklearn.metrics import accuracy_score

CC = accuracy_score(test_target, C)
print(CC)
```
