```python
train = train.drop('Cabin', axis=1)
```

```python
test = test.drop('Cabin', axis=1)
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train['Age'] = train['Age'].fillna(train['Age'].mean())
```

```python
test['Age'] = test['Age'].fillna(test['Age'].mean())
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
target = train.pop('Survived')
```

```python
train_e = train['Embarked']
test_e = test['Embarked']
```

```python
train_e = pd.get_dummies(train_e)
test_e = pd.get_dummies(test_e)
```

```python
pd.DataFrame(train_e)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler() 
train_scaler = ss.fit_transform(train_e)
test_scaler = ss.transform(test_e)
```

```python
pd.DataFrame(train_scaler)
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(train_scaler, target, test_size=0.2, random_state=1204)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, max_depth=9, random_state=1204)
rf.fit(train_input, train_target)
```

```python
A = rf.predict(test_input)
```

```python
from sklearn.metrics import accuracy_score

AA = accuracy_score(test_target, A)
print(AA)
```
